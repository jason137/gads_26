# visualization with python

This notebook illustrates the basic syntax for creating histograms, boxplots, and scatterplots with
[seaborn](http://stanford.edu/~mwaskom/software/seaborn/).
More information is available in the seaborn
[tutorial](http://stanford.edu/~mwaskom/software/seaborn/tutorial.html).

Note that if you're in an IPython session, you'll need to omit the `%` syntax below and include `plt.show()` where indicated with comments to display each plot.

```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## 1. histogram

```python
k = pd.read_csv('../datasets/beer.tsv', sep='\t')
sns.distplot(k.WR, kde=False);
# plt.show()
```

## 2. boxplot

```python
k = pd.read_csv('../datasets/rayleigh.tsv', sep='\t')
k['oxygen'] = k.origin.apply(lambda k: k == 'Air')
ax = sns.boxplot(x='oxygen', y='weight', data=k)
# plt.show()
```

## 3. scatterplot

```python
k = pd.read_csv('../datasets/mauna_loa.tsv', sep='\t')
grid = sns.JointGrid(k.dec_date, k.avg, space=0, size=6, ratio=50)
grid.plot_joint(plt.scatter, color="g")
# plt.show()
```
